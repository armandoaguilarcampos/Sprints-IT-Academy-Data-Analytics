### 1. Calculadora del Indice de Masa Corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades.

La funció ha de classificar el resultat en les seves respectives categories.

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduïdes per l’usuari/a no estan en el format adequat o no pren valors raonables.

```python
def calculadora_IMC():
    '''
    Función para calcular el Indice de Masa Corporal\n
    peso: en kg,
    altura: en m
    '''
    while True:
        try:
            peso = float(input('Indique el peso en kilogramos: ').replace(",", "."))
            if peso > 0 and peso < 200:
                break
            else:
                print('El peso introducido no tiene un valor razonable.')
        except:
            print('El peso introducido no es válido.')

    while True:
        try:
            altura = float(input('Indique la altura en metros: ').replace(",", "."))
            if altura > 0 and altura < 3:
                break
            else:
                print('La altura introducida no tiene un valor razonable.')
        except:
            print('La altura introducida no es válida.')

    IMC = round(peso / (altura**2),2)

    categoría = ''

    if IMC < 18.5:
        categoría = 'Bajo peso'
    elif IMC <= 24.9:
        categoría = 'Peso normal'
    elif IMC <= 29.9:
        categoría = 'Sobrepeso'
    else:
        categoría = 'Obesidad'

    print(f'El IMC tiene el valor {IMC}, esta dentro de la categoría de {categoría}')

    return IMC
```

### 2. Convertidor de temperaturas

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re).

Selecciona almenys 2 conversors, de tal manera que en introduir una temperatura retorni, com a mínim, dues conversions, de manera que es puguin guardar (recorda que un print() no es pot guardar mai).

Consell: Intenta validar les dades prèviament, perquè enviï un missatge d’advertència si les dades introduides per l’usuari/a no estan en el format adequat.

(EXTRA): Pensa una manera d’emmagatzemar totes les possibles conversions en un sol objecte (Llista? Diccionari? DataFrame?) en comptes d’escriure molts if else en funció de la temperatura d’origen i la temperatura de destí.

```python
def convertir_temperaturas():
    '''
    Función para convertir una unidad de temperatura a las otras 2 registradas;  
    regresa diccionario con resultados.\n
    Las unidades de temperatura registradas son:
        Celsius (ºC), Kelvin (K) y Fahrenheit (ºF).\n
    Hay que indicar la unidad de la temperatura introducida y
    posteriormente el valor de la temperatura
    '''

    print('Las unidades de temperatura aceptadas son:')
    print('\t· C: Celsius (ºC)')
    print('\t· K: Kelvin (K)')
    print('\t· F: Fahrenheit (ºF)')

    unidades_registradas = ('C','K','F')

    while True:
        try:
            unidad_inicial = input('Indique la unidad de la temperatura inicial: ').upper()
            if unidad_inicial in unidades_registradas:
                break
            else:
                print('La unidad de temperatura no es válida.')
        except:
            print('El valor introducido no es válido.')

    while True:
        try:
            valor_inicial = float(input('Indique el valor de temperatura: ').replace(",", "."))
            if unidad_inicial == 'C' and valor_inicial >= -273.15:
                break
            elif unidad_inicial == 'K' and valor_inicial >= 0:
                break
            elif unidad_inicial == 'F' and valor_inicial >= -459.67:
                break
            else:
                print('El valor de temperatura no es razonable.')
        except:
            print('El valor no es válido.')

    # Per ºC:
    # K = °C + 273.15
    # ºF = (ºC * 1.8) + 32

    # Per K:
    # ºC = K - 273.15
    # ºF = ((K - 273.15) * 1.8) + 32

    # Per ºF:
    # ºC = (ºF - 32) / 1.8
    # K = ((ºF - 32) / 1.8) + 273.15

    conversiones = {
        "C": {"K": "valor_inicial + 273.15", "F": "(valor_inicial * 1.8) + 32"},
        "K": {"C": "valor_inicial - 273.15", "F": "((valor_inicial - 273.15) * 1.8) + 32"},
        "F": {"C": "(valor_inicial - 32) / 1.8", "K": "((valor_inicial - 32) / 1.8) + 273.15"}
    }

    df_conversiones = pd.DataFrame(conversiones).transpose()

    valores_finales = dict()
    for unidad in unidades_registradas:
        if unidad == unidad_inicial:
            continue
        # print(unidad_inicial,unidad)
        # print(df_conversiones.loc[unidad_inicial,unidad])
        valores_finales[unidad] = round(eval(df_conversiones.loc[unidad_inicial,unidad]),2)
        print(f"{valor_inicial} en {unidad_inicial} equivale a {valores_finales[unidad]} en {unidad}.")

    return valores_finales
```

### 3. Contador de palabras en un texto.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula. Intenta que es gestionin totes les casuístiques possibles que facin que el programa no funcioni correctament.

(EXTRA): Quina és la llargada mitjana de les paraules del text que has escrit? “Hola com va?” hauria de retornar (4+3+2) / 3 = 3

```python
def contador_palabras(texto):
    '''
    Función que cuenta cuantas veces aparece cada palabra en un texto.
    También la longitud promedio de todas las palabras del texto.
    '''
    if isinstance(texto, str):

        contador = dict()
        for palabra in texto.lower().strip().split():
            palabra = palabra.strip()
            # solo consideramos las letras en el string, no signos
            palabra = re.sub(r'[^\w\s]', '', palabra)
            if palabra in contador.keys():
                contador[palabra] += 1
            else:
                contador[palabra] = 1

        print("Veces que aparece cada palabra:")
        pprint.pprint(contador, width=1)

        suma_longitud_palabras = 0
        for key,value in contador.items():
            suma_longitud_palabras += len(key)*value
        promedio_longitud_palabras = round(suma_longitud_palabras/len(contador))

        print(f"La longitud promedio de las palabras del texto es {promedio_longitud_palabras}.")

    else:
        print('El valor introducido no es un string.')
```

### 4. Diccionario inverso (con posibilidad de duplicados)

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviant les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment, juntament amb una llista amb els valors associats a la clau repetida.

```python
def diccionario_inverso(diccionario_original):
    '''
    Función que intercambia las llaves y los valores de un diccionario, 
    creando uno nuevo.
    '''

    print("Diccionario original: ",diccionario_original)

    nuevo_diccionario = dict()
    for key,value in diccionario_original.items():

        if value in nuevo_diccionario.keys():
            print(f"La llave {value} ya existe en el nuevo diccionario.")
            lista_valores = list(nuevo_diccionario[value])
            lista_valores.append(key)
            nuevo_diccionario[value] = lista_valores

        else:
            nuevo_diccionario[value] = key

    print("Nuevo diccionario: ",nuevo_diccionario)
```

```python
diccionario_1 = {"a":1, "b":2, "c":3}
diccionario_2 = {"x":"apple", "y":"banana", "z":"banana", "i": "banana"}
diccionario_inverso(diccionario_1)
print('')
diccionario_inverso(diccionario_2)
```

Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

```python
def contador_limpiador_texto(ruta_texto):
    '''
    Función que cuentas la frecuencia de cada palabra y 
    las agrupa según la letra por cual comienzan.
    '''

    try:
        contador = {}
        # abrimos el archivo
        with open(ruta_texto, "r", encoding="utf-8") as archivo:
            for line in archivo:
                for palabra in line.lower().strip().split():
                    # solo consideramos strings que tienen letras
                    if any(letra.isalpha() for letra in palabra):
                        palabra = palabra.strip()
                        palabra = re.sub(r'[^\w\s]', '', palabra)
                        # verificamos si la primera letra existe
                        if palabra[0] in contador.keys():
                            pass
                        else:
                            contador[palabra[0]] = dict() 
                        # contamos la palabra
                        if palabra in contador[palabra[0]].keys():
                            contador[palabra[0]][palabra] += 1
                        else:
                            contador[palabra[0]][palabra] = 1
                    else:
                        continue
        print("Veces que aparece cada palabra:")
        pprint.pprint(contador, width=1)
    except:
        print('No se ha podido leer el archivo.')
```

### 2. Conversió de tipus de dades.

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir en flotants i l'altra on estan els elements que no es van poder convertir.

```python
def clasificar(elemento,lista_float,lista_otros):
    '''
    Función que clasifica un elemento según este se pueda convertir en float o no.
    '''
    try:
        nuevo_elemento = float(elemento)
        lista_float.append(nuevo_elemento)
    except:
        lista_otros.append(elemento)

    return lista_float,lista_otros

def conversion(lista_inicial):
    '''
    Función que intenta convertir valores a float, regresando dos listas para valores float y para otros.
    '''
    lista_float = []
    lista_otros = []

    for elemento in lista_inicial:
        if isinstance(elemento, (list,tuple)):
            for elemento_2 in elemento:
                lista_float,lista_otros = clasificar(elemento_2,lista_float,lista_otros)
        else:
            lista_float,lista_otros= clasificar(elemento,lista_float,lista_otros)

    return lista_float,lista_otros
```

### 1. Generador de Contrasenyes

En aquest punt, el client ha detectat un problema amb les contrasenyes que utilitzen els seus treballadors. Asdf1234, dates d’aniversari o similars. Per a solucionar-ho ens ha encarregat una funció de Python que generi contrasenyes més segures. La funció ha de dependre dels següents paràmetres:

* longitud (int): Longitud de la contrasenya
* majuscules (bool = True): Si hi ha d’aparèixer majúscules
* minuscules (bool = True): Si hi ha d’aparèixer minúscules
* numeros (bool = True): Si hi ha d’aparèixer números
* signes (bool = False): Si hi ha d’aparèixer caràcters especials (,-$? o similars)

(EXTRA) Explora com podríem fer que la funció copiés la contrasenya automàticament al porta papers de l’ordinador (com si l’haguéssim seleccionat i fet ctrl+copy).

```python

def crear_password(longitud, mayusculas = True, minusculas = True, numeros = True, signos = False):
    '''
    Función que crea contraseñas de cierta longitud y dependiendo de los parámetros indicados.
    '''

    lista_may = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lista_min = lista_may.lower()
    lista_num = "0123456789"
    lista_sig = ",-$%&€@#¿?!¡()[]"

    opciones = [lista_may,lista_min,lista_num,lista_sig]
    numero_opciones = []

    # creamos lista de donde elegimos valores según lo indicado en los argumentos
    if mayusculas:
        numero_opciones.append(0) 
    if minusculas:
        numero_opciones.append(1)
    if numeros:
        numero_opciones.append(2)
    if signos:
        numero_opciones.append(3)

    password = ''

    for _ in range(longitud):
        opcion_next_character = np.random.choice(numero_opciones)
        index_next_character = np.random.randint(len(opciones[opcion_next_character]))
        password += opciones[opcion_next_character][index_next_character]

    #Extra: copiar automáticamente en el portapapeles
    pyperclip.copy(password)
        
    return password
```

### 2. Processament de dades simple

Una companya de feina ens ha demanat un favor, aprofitant que sap que estem aprenent a programar. Té un històric de partits de futbol català en un fitxer, on hi ha emmagatzemat els noms dels equips i els resultats. Necessita que processem les dades de manera automàtica, per extreure’n els resultats que necessita.

Necessita un programa que retorni:
* El nombre total de gols que ha fet cada equip.
* El nom de l’equip més golejador.
* El nom de l’equip més golejat
* La classificació global (cada victòria: 3 pts, empat 1 pts, derrota 0 pts)

```python
def puntos_ganados(goal_made,goal_received):
    '''Función para calcular puntos ganados.'''
    points = 0
    if goal_made > goal_received:
        points = 3
    elif goal_made == goal_received:
        points = 1
    else:
        points = 0
    return points

def resultats_futbol(ruta_partidos):
    '''
    Función que calcula los resultados de un histórico de partidos.
    '''
    try:
        with open(ruta_partidos, "r", encoding="utf-8") as archivo:
            # equipo, goles realizados, goles recibidos, puntos
            df = pd.DataFrame(columns= ['goles_realizados', 'goles_recibidos', 'puntos'])
            for line in archivo:
                values = line.split("\t")
                team1 = values[0]
                team2 = values[2].strip('\n')
                goal1 = int(values[1].split('-')[0])
                goal2 = int(values[1].split('-')[1])
                points1 = puntos_ganados(goal1,goal2)
                points2 = puntos_ganados(goal2,goal1)
                
                if team1 in df.index:
                    df.loc[team1,'goles_realizados'] += goal1
                    df.loc[team1,'goles_recibidos'] += goal2
                    df.loc[team1,'puntos'] += points1
                else:
                    df.loc[team1] = [goal1,goal2,points1]

                if team2 in df.index:
                    df.loc[team2,'goles_realizados'] += goal2
                    df.loc[team2,'goles_recibidos'] += goal1
                    df.loc[team2,'puntos'] += points2
                else:
                    df.loc[team2] = [goal2,goal1,points2]

                df = df.sort_values(by='puntos', ascending=False)

            print('Goles realizados por cada equipo: ')
            print(df['goles_realizados'].sort_values(ascending=False).head())
            print('')
            print('Equipo que ha metido más goles:')
            print(df['goles_realizados'].sort_values(ascending=False).head(1))
            print('')
            print('Equipo que ha recibido más goles:')
            print(df['goles_recibidos'].sort_values(ascending=False).head(1))
            print('')
            print('Clasificación global: ')
            print(df.head())
    except:
        print('No se ha podido leer el archivo.')
```

```python
resultats_futbol("historic partits.txt")
```
